```python
def frame_reduce(i):
  folder_root = f'/content/data/L0{i}_extra/' # thay folder_root
  des_path = f'/content/Videos_L0{i}_reduce'
  os.mkdir(des_path)
  folder_names = sorted(glob.glob(folder_root + '*'))

  for folder_path in folder_names:
      folder_name = folder_path.split('/')[-1]
      new_folder_path = des_path + f'/{folder_name}'
      if not os.path.exists(new_folder_path):
          os.mkdir(new_folder_path)
      print(f'Folder_path: {folder_path}')
      print(new_folder_path)
      count = 0
      for image_name in os.listdir(folder_path):
        image_path = os.path.join(folder_path, image_name)
        image = Image.open(image_path)
        # Resize
        image = image.resize((512, 288))
        # Reduce quality
        image.save(f'{new_folder_path}/'+ image_name, optimize=True, quality=20)
        count += 1
        print(f'Process ...{count}/{len(os.listdir(folder_path))}')
```

```python
!rm -rf /content/Videos_L02_reduce
frame_reduce(2)
```
